```python
# ==============================================================
# ðŸš¢ Titanic ML Project +  Unsupervised Anomaly Detection (Î²-VAE)
# Author: Nabiha Nawshin Shukheemoni
# ==============================================================

# === 1. Import Libraries ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# === 2. Load Dataset (fixed path) ===
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(" Dataset loaded successfully!")
df.head()

# === 3. Basic EDA ===
print("\nDataset Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())

# Simple visualization
sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution")
plt.show()

# === 4. Data Preprocessing ===
# Fill missing numeric values with median
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = df[col].fillna(df[col].median())

# Encode categorical columns
df_encoded = df.copy()
for col in df.select_dtypes(include=['object']).columns:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

print("\nEncoded DataFrame Sample:")
df_encoded.head()

# ==============================================================
#  5. Unsupervised Anomaly Detection using Î²-VAE Autoencoder
# ==============================================================

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler

# Prepare numeric data
numeric_df = df_encoded.select_dtypes(include=[np.number]).dropna()
scaler = MinMaxScaler()
X = scaler.fit_transform(numeric_df)

X_tensor = torch.tensor(X, dtype=torch.float32)
dataset = TensorDataset(X_tensor)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# === Î²-VAE Model Definition ===
class BetaVAE(nn.Module):
    def __init__(self, input_dim, latent_dim=3, beta=0.001):
        super().__init__()
        self.beta = beta
        self.fc1 = nn.Linear(input_dim, 8)
        self.fc_mu = nn.Linear(8, latent_dim)
        self.fc_logvar = nn.Linear(8, latent_dim)
        self.fc2 = nn.Linear(latent_dim, 8)
        self.fc3 = nn.Linear(8, input_dim)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def encode(self, x):
        h = self.relu(self.fc1(x))
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h = self.relu(self.fc2(z))
        return self.sigmoid(self.fc3(h))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decode(z)
        return recon, mu, logvar

# === Loss Function ===
def vae_loss(recon, x, mu, logvar, beta=0.001):
    recon_loss = nn.MSELoss()(recon, x)
    kld = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + beta * kld

# === Initialize & Train ===
model = BetaVAE(input_dim=X.shape[1], latent_dim=3, beta=0.001)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(30):
    total_loss = 0
    for (batch,) in loader:
        recon, mu, logvar = model(batch)
        loss = vae_loss(recon, batch, mu, logvar, beta=0.001)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    if epoch % 5 == 0:
        print(f"Epoch {epoch}: Loss = {total_loss / len(loader):.4f}")

# === Reconstruction Error & Anomalies ===
with torch.no_grad():
    recon, mu, logvar = model(X_tensor)
errors = torch.mean((recon - X_tensor) ** 2, dim=1).numpy()

threshold = np.percentile(errors, 95)
anomalies = numeric_df[errors > threshold]
print(f"\n Detected {len(anomalies)} anomalies out of {len(X)} samples.")

# Plot reconstruction error distribution
plt.figure(figsize=(6,4))
plt.hist(errors, bins=30, color='skyblue', edgecolor='black')
plt.axvline(threshold, color='red', linestyle='--', label='Anomaly Threshold')
plt.title("Reconstruction Error Distribution (Î²-VAE Autoencoder)")
plt.xlabel("Reconstruction Error")
plt.ylabel("Count")
plt.legend()
plt.show()

# Display a few anomalies
print("\nSample of detected anomalies:")
display(anomalies.head())

# ==============================================================
# ðŸ§  6. Summary
# ==============================================================
print("""
 Project Summary:
- Loaded Titanic dataset from web source.
- Performed EDA and preprocessing.
- Implemented Î²-VAE Autoencoder in PyTorch for unsupervised anomaly detection.
- Detected unusual passenger records using reconstruction error.
- Demonstrated key concepts: unsupervised learning, latent representation, and anomaly detection.
""")
```
